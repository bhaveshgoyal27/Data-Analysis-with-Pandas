## The .str.replace() Method 

```python
chicago = pd.read_csv("./data/chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.head(3)
```

```python
chicago["Department"].str.len().astype("int")
```

```python
chicago["Department"] = chicago["Department"].str.replace("MGMNT","Management")
chicago.head(2)
```

```python
chicago["Employee Annual Salary"] = chicago["Employee Annual Salary"].str.replace("$","").astype("float")
```

```python
chicago["Employee Annual Salary"].nlargest(15)
```

## Filtering with string methods

```python
mask = chicago["Position Title"].str.lower().str.contains("water")
chicago[mask]
```

```python
mask = chicago["Position Title"].str.lower().str.startswith("water")
chicago[mask]
```

```python
chicago[chicago["Position Title"].str.lower().str.endswith("inspector")]
```

## More string methods (.strip(), .lstrip(), .rstrip())

```python
chicago["Position Title"] = chicago["Position Title"].str.title().str.strip()
chicago["Name"] = chicago["Name"].str.title().str.strip()
chicago["Department"] = chicago["Department"].str.title().str.strip()
chicago.rename(columns={"Position Title": "Position"}, inplace=True)
chicago.head()
```

## String Methods on Index and Columns

```python
chicago.set_index("Name", inplace=True)
chicago.head()
```

```python
chicago.index = chicago.index.str.strip().str.lower()
chicago.head()
```

```python
chicago.columns = chicago.columns.str.strip().str.upper()
chicago.head()
```

## Split method

```python
chicago = pd.read_csv("./data/chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.head(3)
```

```python
chicago["Name"].str.split(",").str.get(0).str.title().value_counts()
```

```python
chicago["Position Title"].str.split(" ").str.get(0).str.title().value_counts().nlargest(10)
```

## More on splits

```python
chicago["Name"].str.split(",").str.get(1).str.strip().str.split(" ").str.get(0).value_counts()
```

## The `expand` and `n` parameters of `.str.split()` method 

```python
chicago["Name"].str.split(",",expand=True)
```

```python
chicago[["First Name", "Last Name"]] = chicago["Name"].str.split(",",expand=True)
chicago.head()
```

```python
chicago["Position Title"].str.split(" ",expand=True)
```

```python
chicago[["First Title","Last Title"]] = chicago["Position Title"].str.split(" ",expand=True, n=1)
```
