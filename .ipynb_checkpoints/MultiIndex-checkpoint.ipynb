## Create A MultiIndex with the `.set_index()` Method

```python
bigmac = pd.read_csv("./data/bigmac.csv", parse_dates = ["Date"])
bigmac.head(3)
```

## The `.get_level_values()` Method

```python
bigmac = pd.read_csv("./data/bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(20)
```

```python
print(bigmac.index.get_level_values(0))
print()
bigmac.index.get_level_values("Date")
```

```python
print(bigmac.index.get_level_values(1))
print(bigmac.index.get_level_values("Country"))
```

## The `.set_names()` Method on MultiIndex

```python
bigmac = pd.read_csv("./data/bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac.index.set_names(["Date", "Location"], inplace = True)
bigmac.head(10)
```

## Extract Rows from a `MultiIndex DataFrame`

```python
bigmac = pd.read_csv("./data/bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac.loc[("2010-01-01", "Brazil")]
```

## The `.transpose()` Method

```python
bigmac = pd.read_csv("./data/bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac = bigmac.transpose()
bigmac.head(1)
```

## The `.swaplevel()` Method

```python
bigmac = pd.read_csv("./data/bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
bigmac.head(3)
```

```python
bigmac = bigmac.swaplevel()
bigmac.head(3)
```

## The `.stack()` Method

```python
world = pd.read_csv("./data/worldstats.csv", index_col = ["country", "year"])
world.head(3)
```

```python
world.stack()
```

```python
world.stack().to_frame()
```

## The `.unstack()` Method, Part 1

```python
s = world.stack()
s.head(7)
```

```python
s.unstack()
```

```python
s.unstack().unstack().unstack()
```

## The `.unstack()` Method, Part 2

```python
s = world.stack()
s.head(7)
```

```python
s.unstack(0)
```

```python
s.unstack(-1) # or 2 in this case
```

```python
s.unstack("year")
```

## The `.unstack()` Method, Part 3

```python
world = pd.read_csv("./data/worldstats.csv", index_col = ["country", "year"])
s = world.stack()
s.head(3)
```

```python
s.unstack(level = ["year", "country"])
```

```python
s.unstack(level = [0, 1])
```

```python
s = s.unstack("year", fill_value = 0)
```

```python
s.head()
```

## The pivot method

```python
salesmen = pd.read_csv("./data/salesmen.csv", parse_dates = ["Date"])
salesmen["Salesman"] = salesmen["Salesman"].astype("category")
salesmen.head()
```

```python
salesmen.pivot(index = "Date", columns = "Salesman", values = "Revenue")
```

## The `pivot_table()` Method

```python
foods =pd.read_csv("./data/foods.csv")
foods.head(3)
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "min")
```

```python
pd.pivot_table(data = foods, values = "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "min").head(3)
```

## The `pd.melt()` Method

```python
sales = pd.read_csv("./data/quarters.csv")
sales
```

```python
pd.melt(sales, id_vars = "Salesman", var_name = "Quarter", value_name = "Revenue")
```
